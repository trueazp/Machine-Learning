# 1. Import library yang diperlukan

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

# 2. Import dataset

#### Sumber Dataset : https://www.kaggle.com/shazadudwadia/supermarket

```python
df = pd.read_csv("D:\Kuliah\Semester_5\Machine Learning\datasets\GroceryStoreDataSet.csv", names=['products'], sep=',')
df.head()
```

```python
df.shape
```

# 3. Membagi produk kemudian memasukkan ke dalam list "data"

```python
data = list(df["products"].apply(lambda x:x.split(",")))
data
```

# 4. Algoritma Apriori dan Encoding One-Hot

#### Algoritma Apriori untuk mengubah True/False menjadi 1/0. Menggunakan TransactionEncoder, kita konversi daftar menjadi daftar Boolean terenkode. Produk yang dibeli atau tidak dibeli pelanggan selama berbelanja sekarang akan diwakili oleh nilai 1 dan 0.

```python
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data, columns = a.columns_)
df = df.replace(False, 0)
df = df.replace(True, 1)
df
```

# 5. Menerapkan Apriori dan Menampilkan Hasilnya

#### Langkah selanjutnya adalah membuat Model Apriori. Kita dapat mengubah semua parameter dalam Model Apriori dengan package mlxtend. Saya akan mencoba menggunakan parameter min_support untuk pemodelan ini. Untuk ini, saya menetapkan nilai min_support dengan nilai ambang 10%.

```python
df = apriori(df, min_support = 0.1, use_colnames = True, verbose = 1)
df
```

# 6. Menampilkan Support, Confidence, & Lift

### Saya memilih nilai Confidence minimum sebanyak 60%. Dengan kata lain, ketika produk X dibeli, kita dapat mengatakan bahwa pembelian produk Y adalah 60% atau lebih.

```python
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.6)
df_ar
```
